```python
def custom_info(df, mode='all'):
    data_features = df.columns
    data_type = []
    null = []
    null_pct = []
    unique = []
    uniqueSample = []

    for item in data_features :
        data_type.append(str(df[item].dtype))
        null.append(sum(df[item].isnull()))
        null_pct.append(round(df[item].isna().sum()/len(df)*100,2))
        unique.append(df[item].nunique())
        uniqueSample.append(df[item].sample(3).values)

    desc = pd.DataFrame({
        'dataFeatures' : data_features,
        'dataType' : data_type,
        'null' : null,
        'nullPercentage' : null_pct,
        'unique' : unique,
        'uniqueSample' : uniqueSample
    })

    if mode == 'missing_values_only':
        desc = desc[desc.null>0].reset_index(drop=True).drop(['unique', 'uniqueSample'], axis=1)
    return desc

def to_sec(s):
    assert isinstance(s, str)
    r = tuple(int(x) for x in s.split(':'))
    assert len(r) == 3
    hours, mins, secs = r
    return hours*3600 + mins*60 + secs
```

```python
custom_info(df, 'missing_values_only')
```

**Temuan :**
- Hanya ada 3 feature yang rownya berisi null value. Dan ketiganya berkaitan dengan velocity. Persentase null value dari ketiga feature ini lebih dari 66%. Oleh karenanya, feature ini nantinya tidak akan digunakan untuk pembuatan model.

### Feature Engineering

Di bagian ini, saya akan membuat fitur yang isinya berupa konversi waktu (pada entry dan juga exit) dari string ke hitungan detik (int) dan juga selisih waktu antara entry dan exit.

```python
df['time_entry_secs'] = df['time_entry'].apply(to_sec)
df['time_exit_secs'] = df['time_exit'].apply(to_sec)
df['time_delta'] = df['time_exit_secs'] - df['time_entry_secs']
```

Sedangakan dibagian ini, saya menghitung selisih jarak antara entry point dan center point yang ada di daerah kota.

```python
center_of_x = (3750901.5068+3770901.5068)/2
center_of_y = (-19268905.6133-19208905.6133)/2

df['x_entry_delta'] = df['x_entry'].apply(lambda x: x - center_of_x)
df['y_entry_delta'] = df['y_entry'].apply(lambda y: y - center_of_y)
```

Alasan mengapa selisih jarak antara `x_exit` dan `y_exit` ke `center_of_x` dan `center_of_y`, adalah karena `x_exit` dan `y_exit` adalah target. 

### CV Score for a single target

Pada case ini, ada 2 fitur yang dijadikan target, yaitu `x_exit` dan `y_exit`. Namun untuk keperluan komparasi performa yang mana menentukan algoritma yang akan kita gunakan selanjutnya, kita akan menggunakan single target terlebih dahulu, yang mana adalah `x_exit`.

```python
X = df[['x_entry','y_entry', 'time_entry_secs','time_exit_secs','time_delta', 'x_entry_delta', 'y_entry_delta']]
y = df['x_exit']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
# from sklearn.ensemble import RandomForestRegressor // terlalu berat, "Kernel Died"
# from sklearn.ensemble import GradientBoostingRegressor // terlalu lama. lebih dari 50 menit. CV Score = 0.873138
from sklearn.model_selection import cross_val_score

def find_CVS(features,target,model, partition, scoring_system):
    X_train, X_test, y_train, y_test = train_test_split(features,target,random_state=42)
    classification = model()
    score = cross_val_score(classification,features, target, cv=partition, scoring=scoring_system).mean()
    return score

def cv_score_accumulation(algorithm_list, X, y, partition, scoring_system):
    cv_score = []
    notes_arr = []
    for i in algorithm_list :
        score = find_CVS(X,y,i, partition, scoring_system)
        cv_score.append(score)
        print(f"\rQueue: {i}")

    cv_df = pd.DataFrame({
        'Algorithm': algorithm_list,
        'CV Score': cv_score
    })
    return cv_df
```

```python
algorithm_list = [LinearRegression, KNeighborsRegressor, DecisionTreeRegressor, XGBRegressor]
a = cv_score_accumulation(algorithm_list, X, y, 5, 'r2')
a
```

Dengan XGBRegressor yang memberikan CV Score 0.88, maka selanjutnya permodelan akan dibuat berdasarkan XGBRegressor.

**Catatan :**
Semula saya menggunakan 6 algoritma, yaitu LinearRegression, KNeighborsRegressor, DecisionTreeRegressor, GradientBoostingRegressor, RandomForestRegressor, XGBRegressor.

### Train model for multiple target (MultiOutputRegressor)

**Catatan :** Alasan khusus saya menggunakan `MultiOutputRegressor` daripada `RegressorChain` adalah sifat `RegressorChain` yang berantai (bertambah 1 feature dari target secara berulang, sampai dengan target bersisa 1) membuat train proses menjadi lama pada laptop saya yang cukup tua. 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import train_test_split

X = df[['x_entry','y_entry', 'time_entry_secs','time_exit_secs','time_delta', 'x_entry_delta', 'y_entry_delta']]
y = df[['x_exit', 'y_exit']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

# params = {
#     "eta": [0.10, 0.15, 0.3],
#     "max_depth": [ 3, 4, 5, 6, 8],
#     "min_child_weight": [ 1, 3, 5, 7],
#     "gamma":[ 0.0, 0.1, 0.2],
#     "colsample_bytree":[ 0.3, 0.4, 1],
# }

params = {
    "eta": [0.15, 0.3],
    "gamma":[0.0, 0.2],
    "max_depth": [6, 8],
}

gsc = GridSearchCV(
            estimator=XGBRegressor(),
            param_grid=params,
            cv=5, scoring='r2', verbose=1, n_jobs=-1)

grid_result = MultiOutputRegressor(gsc).fit(X_train, y_train)
```
